Section 1: Data and Research Questions Data set

This study is taken from Clare Miln 2021 https://royalsocietypublishing.org/doi/10.1098/rsos.210146. 

In this study, the authors video recorded the sprint swimming performance, competitive ability, and social interactions of zebrafish by placing the zebrafish into different social groups. To be precise, their behavioral patterns and their social hierarchy was ranked based on the competition between the zebrafish within the groups. So, the observations are 150 video recordings of the zebrafish. There are 13 features including FishID, Day, Tank, Length(m), Mass(g), Cstart(m/s), Asprint(m/s), Max realized speed(m/s), Av realized speed(m/s), Aggression, Submission, Socialscore, Compsuccess.

Section 2: Visualizations and Data Summaries

Figure 2: Predictors of Competitive Success

This figure illustrates the relationship between various factors (social score, realized speed, and fast start speed) and competitive success among zebrafish. It is divided into sub-panels:
(a) Scatter plot with data points with Social Score on the x-axis against Competitive Success (number of food items obtained) on the y-axis, indicating that higher social scores correlate with greater success in obtaining food.
(b) Scatter plot with data points with Realized Speed (speed realized by individuals when moving towards a food source within a group of five fish) on the x-axis against Competitive Success, showing a positive correlation.
(c) Scatter plot with Fast Start Speed (speed achieved during a 10 s accelerated spring test (U10)) against Competitive Success, which shows no significant relationship.
(d) Bar plot displays the effect sizes (competitive success) of each predictor (social score, realized speed, U10 speed, and fast start speed) on competitive success, with social score having the largest effect.
This figure supports their hypothesis that social rank (indicated by social score) is a stronger predictor of competitive success than physical sprint capacities (e.g., fast start speed). It visualizes how social behavior and realized speed contribute more significantly to obtaining resources than maximal sprint speed, suggesting the role of social dynamics over pure physical performance in competition. Social hierarchy, rather than raw physical ability, provides zebrafish with a competitive advantage in accessing food resources.


Figure 3: Social Score and Realized Speed Over Time

This figure shows the relationship between Social Score and Realized Speed across three days (Day 1 (red), Day 4 (blue), and Day 7 (green)):
(a) Day 1: Scatter plot shows Realized speed (y-axis) is positively correlated with social score (x-axis), suggesting that initially, faster fish tend to hold higher social ranks.
(b) Day 4: Scatter plot shows that the positive correlation between social score and realized speed continues, indicating that fish with higher social ranks maintain some advantage in speed while competing for resources.
(c) Day 7: Scatter plot shows that the correlation between social score and realized speed diminishes, implying that once hierarchies are fully established, speed becomes less associated with rank.
This figure demonstrates the dynamic nature of the relationship between social hierarchy and realized speed as the social structure solidifies over time. Initially, both physical performance and social rank influence each other, but as hierarchy becomes stable, realized speed loses its predictive power for social score. Figure 3 suggests that while speed is important early in the establishment of social hierarchies, it becomes less relevant as social ranks are solidified which supports the authors hypothesis. The findings highlight that zebrafish may suppress their speed to avoid challenging higher-ranked individuals, showing how social structures influence individual behaviors beyond physical abilities alone. Each plot includes individual data points (each fish group) and linear regression lines to show how the correlation between social score and realized speed changes over time.
